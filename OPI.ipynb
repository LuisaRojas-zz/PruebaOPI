```python
# frecuencia de los delitos
(bd 
  .groupby("delito")
  .agg(frequency=("delito", "count")))
```

```python
tabla1 = bd.groupby(by=['ao_hechos', 'mes_hechos', 'alcaldia_hechos','delito'], as_index=False)['_id'].count()
tabla1
```

```python
tabla1.to_excel('delito_tabla1.xlsx')
```

```python
tabla2 = bd.groupby(by=['delito', 'fiscalia', 'categoria_delito','colonia_hechos', 'alcaldia_hechos'], as_index=False)['_id'].count()
tabla2
```

```python
tabla2.to_excel('delito_tabla2.xlsx')
```

```python
bd['Date'] = pd.to_datetime(bd['fecha_hechos'], infer_datetime_format=True)

bd['year'] = bd['Date'].dt.year
bd['month'] = bd['Date'].dt.month
bd['day_num'] = bd['Date'].dt.day
```

```python
# Agrega el dia de la semana
from datetime import datetime, date, time, timedelta
import calendar

n=len(bd.fecha_hechos)
tupla_diassem = ("Lunes", "Martes", "Miércoles", "Jueves","Viernes", "Sábado","Domingo")
day = []

for i in range(n):
    temp = pd.Timestamp(bd.fecha_hechos[i])
    day.append(tupla_diassem[datetime.isocalendar(temp)[2]-1])
#    print(temp.day_name())

bd['day']=day
```

```python
tabla3 = bd.groupby(by=['delito', 'fiscalia', 'categoria_delito','colonia_hechos', 'alcaldia_hechos', 'ao_hechos', 'mes_hechos'], as_index=False)['_id'].count()
tabla3
```

```python
tabla3.to_excel('delito_tabla3.xlsx')
```

```python
tabla4 = bd.groupby(by=['delito', 'ao_hechos', 'mes_hechos', 'day_num', 'day'], as_index=False)['_id'].count()
tabla4
```
